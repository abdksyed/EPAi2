```
from random import uniform
from math import sqrt

def random_shot(radius):
    rand_x = uniform(-radius, radius)
    rand_y = uniform(-radius, radius)

    if(sqrt(rand_x**2 + rand_y**2) <= radius):
        is_in_circle = True
    else:
        is_in_circle = False

    return rand_x, rand_y, is_in_circle
```

```
count_inside = 0
n = 1_000_000

for i in range(n):
    *_, is_in_circle = random_shot(1)
    if is_in_circle:
        count_inside += 1

print(f'The function gave value = {4 * count_inside / n}')
```

```
class Point3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    def __repr__(self):
        return f"Point3D(x={self.x}, y={self.y}, z={self.z})"
    
    def __eq__(self, other):
        if isinstance(other, Point3D):
            return self.x == other.x and self.y == other.y and self.z == other.z
        else:
            return False
```

```
pt3d_1 = Point3D(10, 20, 30)
pt3d_2 = Point3D(10, 20, 30)
```

```
pt3d_1
```

```
pt3d_1 == pt3d_2
```
